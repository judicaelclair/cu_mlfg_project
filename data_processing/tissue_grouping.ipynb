# Tissue grouping

```python
import pandas as pd
import numpy as np
import json
DATADIR = '../data/'
dname = pd.read_csv(DATADIR+'methylation_processed.csv')
atac = pd.read_csv(DATADIR+'atacseq_processed.csv')
dname = dname.drop(['Unnamed: 0','Unnamed: 0.1'],axis = 1)
atac = atac.drop(['Unnamed: 0','Unnamed: 0.1'],axis = 1)
```

```python
def return_columns(df):
    df = df.drop(['chrom','chromRange','chromEnd','chromStart','gene','strand'],axis = 1)
    return df.columns
```

```python
data_columns = list(return_columns(dname))
```

```python
dname['methylation values'] = list(dname[data_columns].values)
atac['atac values'] = list(atac[data_columns].values)
dname.head()
```

```python
dname_proc = dname.drop(['strand','gene'],axis=1)
atac_proc = atac.drop(['strand','gene'],axis=1)
```

```python
atac_proc  =atac_proc.sort_values(by='chromStart')
dname_proc  = dname_proc.sort_values(by='chromStart')
dname_proc.head()
```

```python
dname_chroms = [x for _, x in dname_proc.groupby(dname_proc['chrom'])]
atacseq_chroms = [x for _, x in atac_proc.groupby(atac_proc['chrom'])]
```

```python
def subseq(seq):
    sub_arrays = np.array_split(seq, len(seq) // 42 + 1)
    sub_list = [list(array) for array in sub_arrays]
    return sub_list

def group_by_tissue(chunks):
    agg_new = {}
    for chunk in chunks:
        chrom = chunk.iloc[0]['chrom']
        chunk = chunk.reset_index(drop = True)
        for column in data_columns:
            subs = np.array_split(np.array(chunk[column]),100)
            for i in range(len(subs)):
                agg_new[chrom+'_'+column+'_'+str(i)] = list(subs[i])

    return agg_new

agg_dname_new = group_by_tissue(dname_chroms)
agg_atac_new = group_by_tissue(atacseq_chroms)
```

```python
with open(DATADIR+'dname_tissue_grouped.json', 'w') as fout:
    json.dump(agg_dname_new, fout)

with open(DATADIR+'atac_tissue_grouped.json', 'w') as fout:
    json.dump(agg_atac_new, fout)
```
